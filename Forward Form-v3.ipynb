```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

path = './Player-data/' 
files = os.listdir(path)
```

```python
for i in files:
    stats = pd.read_csv(path+i, encoding='utf-8')
    player_name = i.split('2')[0]
    stats['npxG+xA'] = stats.npxG+stats.xA
    Filter = stats.Comp=="Premier League"
    stats_xG = stats[Filter][['Round','npxG','xA','npxG+xA', 'Min']]
    
    stats_xG['npxG+xA'] = stats_xG['npxG+xA']*90/stats_xG['Min'].astype(int)
    stats_xG['Week'] = stats_xG['Round'].str.split(' ').str[1]
    stats_xG['Week'] = stats_xG['Week'].astype(int)
    moving_average=[]
    mw = 10
    for i in range(len(stats_xG['npxG+xA'])):
        if i>=mw-1:
            moving_average.append(stats_xG['npxG+xA'].iloc[i-mw+1:i+1].mean())

    plt.figure(figsize = (8,6))
    plt.plot(np.arange(10, len(stats_xG)+1, 1), moving_average, '-o')
    plt.title(player_name)
    x_ticks = []
    plt.xticks(np.arange(10, len(stats_xG)+1, 1), stats_xG['Week'].iloc[9:])
    plt.grid()
    plt.xlabel('Week')
    plt.ylabel('Moving Average of npxG+xA')
    plt.show()
#     break
    
```
